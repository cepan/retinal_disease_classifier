# Data Viz

```python
import pandas as pd
```

```python
train_df = pd.read_csv("../../data/RFMiD/labels/Train.csv")
validation_df = pd.read_csv("../../data/RFMiD/labels/Validation.csv")
test_df = pd.read_csv("../../data/RFMiD/labels/Test.csv")
train_df
```

```python
train_df = train_df[(train_df['Disease_Risk'] == 0) | 
                 ((train_df['Disease_Risk'] == 1) & 
                  ((train_df['DR'] == 1) | 
                   (train_df['MH'] == 1) | 
                   (train_df['ODC'] == 1) | 
                   (train_df['TSLN'] == 1) | 
                   (train_df['DN'] == 1) | 
                   (train_df['MYA'] == 1) | 
                   (train_df['ARMD'] == 1)))]
train_df = train_df[['ID','DR','MH','ODC','TSLN','DN','MYA','ARMD']]
print(train_df.shape)
display(train_df.sum())

validation_df = validation_df[(validation_df['Disease_Risk'] == 0) | 
                 ((validation_df['Disease_Risk'] == 1) & 
                  ((validation_df['DR'] == 1) | 
                   (validation_df['MH'] == 1) | 
                   (validation_df['ODC'] == 1) | 
                   (validation_df['TSLN'] == 1) | 
                   (validation_df['DN'] == 1) | 
                   (validation_df['MYA'] == 1) | 
                   (validation_df['ARMD'] == 1)))]
validation_df = validation_df[['ID','DR','MH','ODC','TSLN','DN','MYA','ARMD']]
print(validation_df.shape)
display(validation_df.sum())

test_df = test_df[(test_df['Disease_Risk'] == 0) | 
                 ((test_df['Disease_Risk'] == 1) & 
                  ((test_df['DR'] == 1) | 
                   (test_df['MH'] == 1) | 
                   (test_df['ODC'] == 1) | 
                   (test_df['TSLN'] == 1) | 
                   (test_df['DN'] == 1) | 
                   (test_df['MYA'] == 1) | 
                   (test_df['ARMD'] == 1)))]
test_df = test_df[['ID','DR','MH','ODC','TSLN','DN','MYA','ARMD']]
print(test_df.shape)
display(test_df.sum())
```

```python
train_df.to_csv("../../data/RFMiD/labels/Filtered_Train.csv", index=False)
validation_df.to_csv("../../data/RFMiD/labels/Filtered_Validation.csv", index=False)
test_df.to_csv("../../data/RFMiD/labels/Filtered_Test.csv", index=False)
```
